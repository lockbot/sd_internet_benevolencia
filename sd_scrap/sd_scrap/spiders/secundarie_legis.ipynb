# Now let's make pizza graphs for each column group

```python
for colname in df.columns[7:14]:
    print(colname)

    s_colstacker = pd.Series(0, index=df.iloc[:, 14:].columns[df.iloc[:, 14:].columns.str.startswith(colname)],
                             dtype='float64')
    for i, row in df.iterrows():
        for col in df.iloc[:, 14:].columns[df.iloc[:, 14:].columns.str.startswith(colname)]:
            s_colstacker[col] += row[col]

    # normalize to get proportions (or percentages)
    total = s_colstacker.sum()
    s_colstacker = s_colstacker / total

    # rename indices
    s_colstacker.index = s_colstacker.index.str.replace(colname + '_', '')

    # display s_colstacker.index but in a column with values in percentage
    display(s_colstacker)
    # plot pie chart
    fig = px.pie(s_colstacker, values=s_colstacker.values, names=s_colstacker.index, title=colname)
    fig.update_layout(template="plotly_white")
    fig.show()
```

```python
# Convert 'data' column to datetime and extract year
df['year'] = pd.to_datetime(df['data'], dayfirst=True).dt.year

# Create 4-year periods
min_year = df['year'].min()
df['period'] = ((df['year'] - min_year) // 4) * 4 + min_year
df['period'] = df['period'].astype(int).astype(str) + '-' + (df['period'] + 3).astype(int).astype(str)

period_counts = df['period'].value_counts().sort_index()

for colname in df.columns[7:14]:
    print(colname)

    df_stack = pd.DataFrame()
    df_count = pd.DataFrame()
    for period in period_counts.index:
        df_part = df[df['period'] == period]

        s_colstacker = pd.Series(0, index=df_part.iloc[:, 14:].columns[
            df_part.iloc[:, 14:].columns.str.startswith(colname)], dtype='float64')
        for i, row in df_part.iterrows():
            for col in df_part.iloc[:, 14:].columns[df_part.iloc[:, 14:].columns.str.startswith(colname)]:
                s_colstacker[col] += row[col]

        # normalize to get proportions (or percentages)
        total = s_colstacker.sum()
        s_colstacker_perc = s_colstacker / total

        # rename indices
        s_colstacker.index = s_colstacker.index.str.replace(colname + '_', '')
        s_colstacker_perc.index = s_colstacker.index

        s_colstacker.name = period
        s_colstacker_perc.name = period
        df_stack = pd.concat([df_stack, s_colstacker], axis=1)
        df_count = pd.concat([df_count, s_colstacker_perc], axis=1)

    # plot stacked bar chart
    df_stack = df_stack.reset_index().melt(id_vars='index', var_name='period', value_name='leis')
    df_count = df_count.reset_index().melt(id_vars='index', var_name='period', value_name='percentage')
    df_plot = df_stack.merge(df_count)
    fig = px.bar(df_plot, x='period', y='leis', color='index', title=colname,
                 hover_data=['percentage'], barmode='stack')
    fig.update_layout(template="plotly_white")
    fig.show()
```

```python
benevolencia_cols = [{'name': 'benevolencia_benefício', 'color': 'green'},
                     {'name': 'benevolencia_ameaça', 'color': 'red'}, {'name': 'benevolencia_neutra', 'color': 'grey'}]

for colname in df.columns[[7] + list(range(9, 14))]:
    print(colname)

    df_layers = pd.DataFrame()
    for nc in benevolencia_cols:
        df_layers[nc['name']] = pd.Series(0, index=df.iloc[:, 14:].columns[
            df.iloc[:, 14:].columns.str.startswith(colname)], dtype='float64')
    for i, row in df.iterrows():
        for col in df.iloc[:, 14:].columns[df.iloc[:, 14:].columns.str.startswith(colname)]:
            for nc in benevolencia_cols:
                df_layers.loc[col, nc['name']] += row[col] * row[nc['name']]

    # normalize to get proportions (or percentages)
    total = df_layers.sum().sum()
    df_layers_perc = df_layers / total

    # rename indices
    df_layers_perc.index = df_layers_perc.index.str.replace(colname + '_', '')

    # rename columns
    df_layers_perc.columns = df_layers_perc.columns.str.replace('benevolencia_', '')

    fig = go.Figure()
    for nc in benevolencia_cols:
        fig.add_trace(go.Bar(
            x=df_layers_perc[nc['name'].replace('benevolencia_', '')],
            y=df_layers_perc.index,
            name=nc['name'].replace('benevolencia_', ''),
            marker_color=nc['color'],
            orientation='h',
        ))
    # plot pie chart
    fig.update_layout(title=colname, template="plotly_white")
    fig.update_xaxes(tickformat='.0%')
    fig.show()
```

```python
for colname in df.columns[[7, 12, 13]]:
    # Create subplot figure with 1 row and the same number of columns as periods
    fig = make_subplots(rows=1, cols=len(period_counts), subplot_titles=period_counts.index)

    print(colname)

    # Loop over periods
    for col_i, period in enumerate(period_counts.index, start=1):
        df_sub = df[df['period'] == period]
        df_layers = pd.DataFrame()
        for nc in benevolencia_cols:
            df_layers[nc['name']] = pd.Series(0, index=df_sub.iloc[:, 14:].columns[
                df_sub.iloc[:, 14:].columns.str.startswith(colname)], dtype='float64')
        for i, row in df_sub.iterrows():
            for col in df_sub.iloc[:, 14:].columns[df_sub.iloc[:, 14:].columns.str.startswith(colname)]:
                for nc in benevolencia_cols:
                    df_layers.loc[col, nc['name']] += row[col] * row[nc['name']]

        # normalize to get proportions (or percentages)
        total = df_layers.sum().sum()
        df_layers_perc = df_layers / total

        # rename indices
        df_layers_perc.index = df_layers_perc.index.str.replace(colname + '_', '')

        # rename columns
        df_layers_perc.columns = df_layers_perc.columns.str.replace('benevolencia_', '')

        for nc in benevolencia_cols:
            fig.add_trace(go.Bar(
                x=df_layers_perc.index,
                y=df_layers_perc[nc['name'].replace('benevolencia_', '')],
                name=nc['name'].replace('benevolencia_', ''),
                marker_color=nc['color'],
            ), row=1, col=col_i)  # Add the bar chart to the subplot

    # Update layout and axes
    fig.update_layout(title=colname, template="plotly_white")
    fig.update_yaxes(tickformat='.0%')
    # Show the figure
    fig.show()
```
